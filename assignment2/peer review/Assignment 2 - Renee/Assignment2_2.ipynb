```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
from retinaface import RetinaFace
```

```python
# return facial area coordinates and landmarks (eyes, nose, and mouth) with confidence scores
resp = RetinaFace.detect_faces("sportsteam.png")

# Extract faces and apply alignment
faces = RetinaFace.extract_faces(img_path="sportsteam.png", align=True)

# Display the aligned faces
for face in faces:
    plt.imshow(face)
    plt.show()
```

```python
# Extract relevant information (facial area coordinates and landmarks)
face_info = resp["face_1"]
confidence_score = face_info["score"]
facial_area = face_info["facial_area"]
landmarks = face_info["landmarks"]

print(f"Confidence Score: {confidence_score:.4f}")
print(f"Facial Area Coordinates: {facial_area}")
print(f"Landmarks (Right Eye, Left Eye, Nose, Mouth): {landmarks}")
```
